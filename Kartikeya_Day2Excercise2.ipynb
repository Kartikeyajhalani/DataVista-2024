```python
#Q1. Find the car with the best mpg.
```

```python
print('Car with the best mpg is')
ans = df[df['mpg'] == df['mpg'].max()]
print(ans['model'])
```

```python
#Q2. Find the car with the worst mpg.
```

```python
print('Car with the worst mpg is')
ans = df[df['mpg'] == df['mpg'].min()]
print(ans['model'])
```

```python
#Q3. Find the car with the best horsepower.
```

```python
print('Car with the best HP is')
ans = df[df['hp'] == df['hp'].max()]
print(ans['model'])
```

```python
#Q6.  What is average mpg for manual vs. automatic cars ?
```
